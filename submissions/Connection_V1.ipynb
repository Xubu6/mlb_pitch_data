```python
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
import mysql.connector
```

```python
connect = mysql.connector.connect(user='root', password='root',
                              host='127.0.0.1', port = 8889,
                              database='pitch_data')
```

```python
curse = connect.cursor()
```

```python
curse.execute("select * from atbats limit 1000")
result = curse.fetchall
```

```python
ab_ids = []
batter_ids = []
events = []
g_ids = []
innings = []
os = []
p_scores = []
p_throwss = []
pitcher_ids = []
stands = []
tops = []
 
for i in curse:
    ab_ids.append(i[0])
    batter_ids.append(i[1])
    events.append(i[2])
    g_ids.append(i[3])
    innings.append(i[4])
    os.append(i[5])
    p_scores.append(i[6])
    p_throwss.append(i[7])
    pitcher_ids.append(i[8])
    stands.append(i[9])
    tops.append(i[10])
     
print("ab_ids: ", ab_ids)
print("batter_ids: ", batter_ids)
print("g_ids: ", g_ids)
print("innings: ", innings)
     
print("os: ", os)
print("p_scores: ", p_scores)
print("p_throwss: ", p_throwss)
print("pitcher_ids: ", pitcher_ids)

print("stands: ", stands)
print("tops: ", tops)
```
